```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def four_true(k,t,n,c):
    return (np.sqrt(k+t)/(1 - np.sqrt((c*k*t)/n)))*np.sqrt( (c*(max([k,t]))  )/n  )

def four(k,n,c):
    a = ((n/c) - (k**2))**2
    b = (c/(n*k))
    return a*(1/4.0)*b

def book(k,n,c):
    return (np.sqrt(n/(c*41))*4 - 2*k  )*0.5 

def rns(k,t,n,c):
    return ( np.sqrt(k+t)*(1.0/np.sqrt(k) + 1.0/np.sqrt(t))*np.sqrt((4*c*k*t)/n))
```

```python
n_1 = 125*125
n_2 = 32*32
n_3 = 75*75
```

# N=125*125

```python
#Lets get the t-values from THeorem 4.33. We feed k, n, and c, and see what is the largest t that the theorem will
#allow. This happens when theta = 1. 
n1_x,n1_t_four = [],[]
for i in range(40,88):
    n1_t_four.append(four(float(i),float(n_1),2.0))
    n1_x.append(i)
```

```python
plt.plot(n1_x,n1_t_four,'b-')
#plt.axis([0,np.sqrt(n_1),0,10])
plt.xlabel('k-values')
plt.ylabel('t-values')
plt.title('Theorem 4.33 : N=125*125')
```

```python
#Lets see what t values the (k+t)-sparse approach allows. Here we set \delta_{2k+2t}: (2k + 2t)sqrt(c/n) = 1
#and get t values, given k, n, and c values.
n1_t_book = []
for i in n1_x:
    n1_t_book.append(book(float(i),float(n_1),2.0))
```

```python
plt.plot(n1_x,n1_t_book,'r-')
#plt.axis([0,np.sqrt(n_1),0,10])
plt.xlabel('k-values')
plt.ylabel('t-values')
plt.text(50,-50,'t < 0, hence no budget')
plt.title('(K+t)-Sparse from Book : N=125*125')
```

```python
#Modified Robust Null Space
n1_rho=[]
for k,t in zip(n1_x,n1_t_four):
    n1_rho.append(rns(float(k),float(t),float(n_1),2.0))
```

```python
#Lets see if the (k,t) values that work for Theorem 4.33, work for the modified Robust Null Space proof. 
#Here we feed the valus of (k,t) that were allowed by Theorem 4.33 and check wether the constant
# sqrt(k+t)* (1/sqrt(k) + 1/sqrt(t)) * rho < 1. If the constant (y-axis) is not less than 1, then the we cannot use 
#the (k,t) pair.
plt.plot(n1_x,n1_rho,'b-')
plt.xlabel('k-values')
plt.text(50,2.15,'y should be <= 1, but isnt.')
plt.ylabel('sqrt{k+t}*c*rho')
plt.title('Modified Robust Null Space : N=125*125')
```

# 4.33 vs RNS Analysis

## N = 125*125

```python
#Lets fix n = 125*125 and see which values of (k,t) give us the constant < 1.
t_x, t_y, f_x, f_y = [],[],[],[]
for i in range(75):
    for j in range(75):
        if (four_true(float(i),float(j),float(n_1),2.0) <=1):
            t_x.append(i)
            t_y.append(j)
        else:
            f_x.append(i)
            f_y.append(j)
```

```python
plt.scatter(t_x,t_y,c='green')
plt.scatter(f_x,f_y,c='red')
plt.xlabel('k-values')
plt.text(35,2.16,'Green are good, Red are bad')
plt.ylabel('t-values')
plt.title('4.33 : N=125*125')
```
